```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("corona_gucci_20210523.xlsx")
```

```python
df.head()
```

```python
df['Date'] = df['Date'].replace(to_replace=r'^.*:.*$', value='22.05.2021', regex=True)
```

```python
swiat = df[df["Location"] == "Świat"]
polska = df[df["Location"] == "Polska"]
```

```python
x = pd.to_datetime(swiat['Date'], dayfirst = True).dt.date
y = swiat['Mean_sentiment']
plt.scatter(x, y, s = 10, label='świat')

x = pd.to_datetime(polska['Date'], dayfirst = True).dt.date
y = polska['Mean_sentiment']
plt.scatter(x, y, s = 10, label='polska')

plt.legend()
plt.xticks(rotation=90)
plt.yticks()
plt.xlabel('Publication date', labelpad = 10)
plt.ylabel('Sentiment')

plt.title('TVP INFO',  pad = 20)


plt.show()
```
